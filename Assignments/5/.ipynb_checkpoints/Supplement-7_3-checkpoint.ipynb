### 7.3 Programming Task: Color Quantization using K-Means clustering.

 The goal of this task is to perform color quantization using the K-Means clustering algorithm.

   i\. Implement the K-Means clustering algorithm using Numpy/SciPy.




```python
def initialize_centroids(X, k):
    """ K-Means++ Initialization: Pick first centroid randomly, then pick next centroids based on distance. """
    centroids = [X[np.random.randint(len(X))]]
    for _ in range(1, k):
        distances = np.array([min(np.linalg.norm(x - c) for c in centroids) for x in X])
        probabilities = distances / distances.sum()
        new_centroid = X[np.random.choice(len(X), p=probabilities)]
        centroids.append(new_centroid)
    return np.array(centroids)

def assign_clusters(X, centroids):
    """ Assign each point to the nearest centroid """
    distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

def update_centroids(X, labels, k):
    """ Compute new centroids and handle empty clusters """
    new_centroids = np.zeros((k, X.shape[1]))
    for i in range(k):
        if np.any(labels == i):
            new_centroids[i] = X[labels == i].mean(axis=0)
        else:  # If a cluster is empty, reinitialize it with a random point
            new_centroids[i] = X[np.random.randint(len(X))]
    return new_centroids

def K_means_clustering(train_X, k=64, max_iters=100, tol=1e-4):
    """
    Implements the K-Means clustering algorithm with K-Means++ initialization.
    
    Parameters:
    - train_X: Input array of shape (p, 3), where p is the number of pixels.
    - k: Number of clusters (default: 64).
    - max_iters: Maximum number of iterations (default: 100).
    - tol: Tolerance for convergence (default: 1e-4).
    
    Returns:
    - centroids: The final cluster centers.
    - labels: The cluster assignments for each point.
    """
    X = np.array(train_X)
    centroids = initialize_centroids(X, k)

    for _ in range(max_iters):
        labels = assign_clusters(X, centroids)
        new_centroids = update_centroids(X, labels, k)

        # Check for convergence
        if np.linalg.norm(new_centroids - centroids) < tol:
            break
        
        centroids = new_centroids

    return centroids, labels
```

   ii\. Read the given image. Consider each pixel as a 3-dimensional vector and run your K-Means clustering algorithm to get 64 clusters.


```python
# Read image
image_path = "sup_7_3_input.jpg"  # Replace with the actual image path
image = imread(image_path)
image = image / 255.0  # Normalize to [0, 1]

# Reshape image into a list of RGB pixels
pixels = image.reshape(-1, 3)

# Apply K-Means Clustering
k = 64  # Number of clusters
centroids, labels = K_means_clustering(pixels, k)

# Replace each pixel with its corresponding centroid
quantized_pixels = centroids[labels]
quantized_image = quantized_pixels.reshape(image.shape)

# Display original and quantized image
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(quantized_image)
axes[1].set_title("Quantized Image (K=64)")
axes[1].axis("off")

plt.show()

# Save the quantized image, ignoring the low contrast warning
imsave("quantized_image.jpg", (quantized_image * 255).astype(np.uint8), check_contrast=False)
```
